```python
#problem 
# input = "today has been great for me"
#output  = "Today Has Been Great For Me"


input_str = "today has been great for me"
listss = []
input_str.split()
for i in input_str.split():
    i.capitalize()
    listss.append(i.capitalize())

print(listss)
print(" ".join(listss))
```
