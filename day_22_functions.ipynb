```python
# Components of a function
# 1. Function Definition
# 2. Function Call
# 3. Function Arguments
# 4. Function Parameters
# 5. Return Statement

def greet(name):               # Function Definition
    """This function greets the person whose name is passed as parameter"""
    print("Hello, " + name + ". Good morning!")  
    return "Greeted " + name      # Return Statement    
```

```python
greet("Rishabh")
```

```python
def is_even(number):
    """This function checks if a number is even or not"""
    if type(number) == int:
        if number % 2 == 0:
            return True
        else:
            return False
    else:
        return "Invalid Input"
result = is_even(4)       

print(result) # Function Call
```

Parameters and Arguments


1. Default Arguments
2. Positional Arguments
3. Keyword Arguments
4. Arbitrary Arguments

```python
# Parameters vs Arguments
# Parameters are the variables listed inside the parentheses in the function definition
# Arguments are the values passed to the function when it is called
def add(a, b):          # a and b are parameters
    return a + b
sum = add(5, 10)       # 5 and 10 are arguments
print(sum)
```

Default arguments

```python
def power(a = 1 , b = 2):   # default parameters
    return a ** b

number = power() # Function Call with no arguments
print(number)
number = power(3 , 3) # Function Call with positional arguments
print(number)
number = power(b = 4) # Function Call with keyword argument
print(number)
number = power(a = 2) # Function Call with keyword argument
print(number)
number = power(b = 5 , a = 2)   # Function Call with keyword arguments in different order
print(number)
number = power(4) # Function Call with one positional argument
print(number)
```

```python
# Arbitrary Arguments
def multiply(*args):   # *args is used to pass arbitrary number of positional arguments
    result = 1
    for num in args:
        result *= num
    return result
```

```python
def flexi(*number):
    product = 1
    for i in number:
        product *= i
    print(product)
    
# in this number of arguments can be variable and stored in tuple named 'number'
```

```python
flexi(1, 2, 3)  # Function Call with 3 arguments
flexi(4, 5)     # Function Call with 2 arguments
```

```python
# Nested functions

def f():
    print("Inside function f")
    def h():
        print("Inside nested function h")
        
    h()  # Calling nested function h inside function f
    
    
```

```python
# Example of returning a function from another function (closure)
def f():
    def x(a,b):
        return a + b
    return x

val = f()(3,4)  # Calls f(), which returns x, then calls x(3,4)
print(val)      # Output: 7
```
