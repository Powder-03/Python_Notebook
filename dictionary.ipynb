```python
# Dictionary
# Dictionary has no indexing and slicing
# It is a mutable datatype
# Keys are always unique and immutable
# Values can be duplicate and mutable or immutable
```

```python
# Mutables ->  list, set, dictionary (changes will be on same memory address)
# Immutables ->  int, float, string, tuple (changes will be on different memory address)
```

```python
# Create
d = {
    "name": "Rishabh",
    "age": 21,
    "city": "New Delhi"}
print(d)  # {'name': 'Rishabh', 'age': 21, 'city': 'New Delhi'}
```

```python
d1 = {(1.,2): "Nitish"}
print(d1)
```

```python
d2 = {"name": "Rishabh", "name": "Nitish"}
print(d2)  # {'name': 'Nitish'}  # duplicate keys not allowed
```

```python
D3 = {"Name":"Rishabh", "Marks":{"Maths":90, "Science":85}}
print(D3)  # {'Name': 'Rishabh', 'Marks': {'Maths': 90, 'Science': 85}}
```

```python
D3["Name"]
```

```python
# Edit in dictionary
D3["Name"] = "Nitish"
print(D3)  # {'Name': 'Nitish', 'Marks': {'Maths': 90, 'Science': 85}}
```

```python
D3.get("Name") # it is beneficial when key is not present it will return None instead of error
```

```python
# Add new key value pair
D3["Age"] = 21
print(D3)  # {'Name': 'Nitish', 'Marks': {'Maths': 90, 'Science': 85}, 'Age': 21}
```

```python
D3['Marks']['Maths']= 95
print(D3)  # {'Name': 'Nitish', 'Marks': {'Maths': 95, 'Science': 85}, 'Age': 21}
```

```python
del D3['Age']
print(D3)  # {'Name': 'Nitish', 'Marks': {'Maths': 95, 'Science': 85}}
```

```python
# using command d3.clear() we can clear all items in dictionary
```

```python
# we can use loops in dictionary as well
for key in D3:
    print(key, ":", D3[key])    
    
```

```python
# wwe iterate through keys and values using items() method
for key, value in D3.items():
    print(key, ":", value)
```
