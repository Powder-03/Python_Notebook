```python
for i in range(1,11):
    if i==5:
        break
    print(i)
```

```python
for i in range(1,11):
    if i==5:
        continue
    print(i)
    
# in continue statement the loop will skip the current iteration and move to the next iteration
# real world example of continue statement is suppose we are showing product on ecommerce website and some product is out of stock so we can use continue statement to skip that product and show the next product
```
