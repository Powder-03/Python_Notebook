```python
# indexing and Slicing
```

```python
c = 'hello'
print(c)
```

```python
print(c[0])  # first character
```

```python
print(c[-1])
```

```python
#string support two types of indexing positive and negative indexing
# positive indexing starts from 0 to n-1
# negative indexing starts from -1 to -n
```

```python
# Slicing 
c = 'Hello, World!'
print(c[0:5])  # 'Hello'
```

```python
print(c[0:5:2])  # 'Hlo' (every second character)
```

```python
print(c[2:])
```

```python
print(c[::-1])  # 'dlroW ,olleH' (reversed string)
```

```python
c[0] = 'X'  # This will raise an error because strings are immutable
```

```python
#strings are immutable means we cannot change the individual characters of a string once it is created.
```

```python
del c
# string variable c is deleted now trying to print c will raise an error
print(c)
```
