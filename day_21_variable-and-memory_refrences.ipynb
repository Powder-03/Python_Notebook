```python
# everything that happens when program is running it is in RAM
# In RAM data is stored in registers which are made up of flip flops
# flip flops can store 1 bit of data either 0 or 1
# flipflop are made up by logic gates and logic gates are made up of transistors

```

```python
# When we store something things are stored in memory and refrenced with address (hexadecimal number)
```

```python
# python calls variable as name 
# name is a reference to the memory location where data is stored
# when we do a assignment like a = 5
# here 5 is stored in some memory location and a is reference to that memory location
# when we do b = a
# here b is also reference to the same memory location where 5 is stored
# so both a and b point to same memory location
# when we do a = 10
# here a is now reference to new memory location where 10 is stored
# but b is still reference to memory location where 5 is stored
# we can check memory location using id() function
```

```python
a = 4
print(id(a))  # memory location of 4 and this pointing is called call by object refrence
b = a
print(id(b))  # memory location of 4
a = 10
print(id(a))  # memory location of 10
```

```python
# in python everything is an object so we call it as call by object refrence
```

```python
# Aliasing 
# if a= 4
# and b = a
# here both a and b point to same memory location
```

```python
import sys
idli= 100
print(sys.getrefcount(idli))  # gives number of references to the object 100
```

```python


# Garbage collection
# when there is no reference to an object in memory that object is called garbage
# python has an inbuilt garbage collector which cleans up the garbage objects from memory
# this is done using reference counting
# when reference count of an object becomes zero that object is garbage and is cleaned up by garbage collector
import gc
gc.collect()  # this will run the garbage collector
# output will be number of unreachable objects it has collected
```

```python
a = 4
b = 4 
# this is not aliasing but both a and b point to same memory location because of python memory optimization for small integers
# this behaviour is only for small integers from -5 to 256
print(id(a))
print(id(b))
```

```python
a = 257
b = 257
# here both a and b point to different memory location because python optimizes memory only for small integers
print(id(a))
print(id(b))
```

```python
a = "haldia_institute"
b = "haldia_institute"
# here both a and b point to same memory location because of python memory optimization for strings and both are valid identifiers
# contains only alphabets , numbers and _ and does not start with number
print(id(a))
print(id(b))

a = "haldia institute"
b = "haldia institute"
# here both a and b point to different memory location because of space in string
print(id(a))
print(id(b))
```

```python
L = [1,2,3]
# it is stored in some memory location
print(id(L))
print(id(L[0]))
print(id(L[1]))
print(id(L[2]))
```

```python
# Mutability refers to the ability to change or edit data in its memory location
# mutability depends on data type
# mutable data types : list , dict , set
# immutable data types : int , float , str , tuple
```

```python
a = "hello"
print(id(a))
a = a +"world"
print(id(a))
# changes happened but changes are in new memory location
```

Side effects of mutability

```python
 l = [1 , 2 , 3]
 l1 = l
 # both will have same memory location
 l1.append(4)
 # now changes have been made in the originial list as both l and l1 point to same memory location
 # suppose we are junior developer and senior told us to do something we need data of l so we did aliasing and do our operation
 # and when senior dev wok back on his variables he found his code is broken because we changed the original list
    # so to avoid this we can use copy module to create a copy of the list
```

```python
# so we do is
L1 = l[:]
print(id(L1))
print(id(l))   
# now both copy are at different memory location and changes made in one will not affect other
```
