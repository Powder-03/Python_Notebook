```python
# These functions are available for strings only
```

```python
c.capitalize() #convert     first character to uppercase
```

```python
"its a rainy day".title() #convert first character of each word to uppercase
```

```python
c.upper()
```

```python
c.lower() #convert all characters to lowercase
```

```python
"KolKaTa".swapcase() #convert uppercase to lowercase and vice versa
```

```python
"it is raining".count("i") #count occurrences of a substring in a string
```

```python
"it is raining".find("i") #find the first occurrence of a substring in a string, returns -1 if not found
```

```python
"it is raining".index("i") #find the first occurrence of a substring in a string, raises an error if not found
```

```python
"is is raining".endswith("g") #check if the string ends with a particular substring
```

```python
"it is raining".startswith("i") #check if the string starts with a particular substring
```

```python
"Hello my name is {} and I am {}".format("Rishabh", 20) #string formatting using format() function
```

```python
"Hello my name is {1} and I am {0}".format(20, "Rishabh") #string formatting using positional arguments
```

```python
"FLAT20".isalnum() #check if all characters in the string are alphanumeric
```

```python
"Flat20@".isalpha() #check if all characters in the string are alphabetic
```

```python
"Who is the pm of India".split() #split the string into a list of words based on whitespace
```

```python
"Who is the pm of India".split("pm") #split the string into a list of words based on whitespace
```

```python
#join is reverse of split
"-".join(["2024", "06", "15"]) #join the list of strings into a single string with "-" as separator
```

```python
"HI my name is Rishabh".replace("Rishabh", "Rishu") #replace occurrences of a substring with another substring
```

```python
name = "  Rishabh  "
name.strip() #remove leading and trailing whitespace
```
