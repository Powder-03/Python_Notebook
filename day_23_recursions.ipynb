```python
# code to multiply two numbers
def multiply(a,b):
    
    if b==1:
        return a
    
    else:
        return a + multiply(a,b-1)
    
result = multiply(5,3)
print(result)
```

```python
import time
def fact(num):
    if num ==1:
        return 1
    else:
        return num*fact(num-1)
start = time.time()    
print(fact(50))
print( time.time() - start)
```

```python
# check palindrome using recursion
def is_palindrome(s):
    if len(s) <= 1:
        return True
    else:
        if s[0] != s[-1]:
            return False
        else:
            return is_palindrome(s[1:-1])
        
        
print(is_palindrome("radar"))
```

```python
# the rabbit problem
def rabbit_pairs(n):
    if n == 1 or n == 2:
        return 1
    else:
        return rabbit_pairs(n-1) + rabbit_pairs(n-2)
    
print(rabbit_pairs(6))
```
