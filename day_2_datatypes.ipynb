Python support 3 categories of data types:

1. Basic types - (integer , float , complex,boolean and string)
2. Container types - (List,Tuple,Sets and Dictionary)
3. User-defined types - Class

```python
#integer 
print(4)
#print function does not get overflowed like other programming languages
#it gets overflowed only when we print 1e309
print(1e309)  #prints inf
```

```python
#float
print(4.5)
```

```python
#boolean
print(True)
print(False)
```

```python
#complex
print(3+4j)
```

```python
#string
print("hello world")
print('hello')
print("""hello world""")
```

```python
#list
print([1, 2, 3, 4])
```

```python
#tuple
print((1, 2, 3, 4))
```

```python
#sets
print({1, 2, 3, 4})
```

```python
#dict
print({"name": "Alice", "age": 25})
```
