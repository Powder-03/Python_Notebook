Literal is a raw data given in a variable. In python, there are various types of literals they are as follows:
1. Numeric Literals 
2. String literals
3. Boolean Literals
4. Special Literals

1. Numeric Literals


```python
a = 0b1010 # binary representation of 10
b = 100
c = 0o12  # octal representation of 10
d = 0xA   # hexadecimal representation of 10

#float literals
float_1 = 10.5
float_2 = 1.5e2  # scientific notation for 150.0
float_3 = 2.5E-3  # scientific notation for 0.

#complex literals
complex_1 = 3 + 4j
complex_2 = complex(5, -6)
x= 3.14j

print(a, b, c, d, float_1, float_2, float_3, complex_1, complex_2)
print(x, x.imag , x.real)
```

2. String literals

```python
string = 'This is python'
strings = "This is python"
char = 'A'
multiline_string = '''This is a multiline
string in python'''
unicode_string = u'This is a unicode string: \u03A9'  # Omega symbol
raw_string = r'This is a raw string: \n no new line'
print(string)
print(strings)  
print(char)
print(multiline_string)
print(unicode_string)
print(raw_string)
```
