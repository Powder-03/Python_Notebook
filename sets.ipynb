```python
# but we can add items in a set using add() method
s5 = {1,2,3}
print(id(s5))
s5.add(4)
print(s5)
id(s5)
```

```python
# deleting items from set using remove() and discard()
s5.remove(3)
print(s5)
```

```python
s5.pop() # removes arbitrary item
```

```python
s1 = {1,2,3,4}
s2 = {3,4,5,6}
#union
print(s1 | s2)
print(s1.union(s2))
#intersection
print(s1 & s2)
print(s1.intersection(s2))

# difference
print(s1 - s2)
print(s1.difference(s2))
```

```python
s1.isdisjoint(s2)
s1.issubset(s2)
s1.issuperset(s2)
s1.copy()
```
